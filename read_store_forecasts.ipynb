```r
library(covidHubUtils)
```

```r
# Load forecasts that were submitted in a time window from zoltar
inc_case_targets <- paste(1:4, "wk ahead inc case")
forecasts <- covidHubUtils::load_latest_forecasts(models = "COVIDhub-ensemble",
                      last_forecast_date = Sys.Date(),
                      forecast_date_window_size = 8,
                      locations = "US",
                      types = c("point","quantile"),
                      targets = c(inc_case_targets),
                      source = "zoltar",
                      hub_repo_path = "")
```

```r
# Plot point forecasts with prediction intervals and without user-provided truth data
p <- covidHubUtils::plot_forecast (forecast_data = forecasts,
                                   models = "COVIDhub-ensemble",
                                   target_variable = "inc case",
                                   locations = "US",
                                   intervals = c(.5, .8, .95),
                                   horizon = 4,
                                   truth_source = "JHU",
                                   plot = TRUE,
                                   truth_as_of = NULL,
                                   title = "default", 
                                   show_caption = TRUE)
```

# Read incident deaths and store them in specified path

```r
target_end_dates <- lubridate::ymd(20200504) + seq(7, 7*40, 7)
inc_case_targets <- paste(0:28, "day ahead inc death")

forecasts <- covidHubUtils::load_latest_forecasts(
                      last_forecast_date = target_end_dates,
                      types = c("point","quantile"),
                      locations = "US",
                      targets = c(inc_case_targets),
                      source = "zoltar",
                      hub_repo_path = "")

path_to_save = '/Users/chaosdonkey06/Dropbox/BIOMAC/EnsembleForecast'
write.csv(forecasts, file.path(path_to_save, paste("forecast_deaths_all", ".csv", sep="")), row.names=FALSE)
```
